```python
#Data Pre-Processing Module

import re
import nltk
nltk.download('punkt')
def remove_nonEnglish(data):
    #words_list=set(nltk.corpus.words.words())
    data_final=""
    words=nltk.word_tokenize(data)
    for word in words:
        if(word in words):
            s=""
            for i in range(0,len(word)):
                if((ord(word[i])>=97 and ord(word[i])<=122) or (word[i] is '.')):
                    s=s+word[i]
            if(len(s)>0):
                data_final=data_final+s+" "

    data_final=re.sub(r'\s\.','.',data_final)       
    return data_final;
    
def preprocess():
    with open("data_set.txt",'r',encoding='utf8') as file2:
        data=file2.read()
    #To remove data between () and []
    data=re.sub("[\(\[].*?[\)\]]", "", data)
    #To remove numbers, percentages and degrees
    data=re.sub(r'\d+(nd|th|st|rd)?|\d+(\.\d*)?% | \d+.C','',data)
    #to remove links
    data=re.sub(r'^(https?:\/\/)?([\da-z\.-]+)\.([a-z\.]{2,6})([\/\w \.-]*)*\/?$','',data)
    data=re.sub(r'[-_,\/\\!@#$%^&*><;:\|"]',' ',data)
    data=re.sub(r'\.','. ',data)
    data=re.sub(r'(\s\s+)',' ',data)
    data=re.sub(r'\.\.+','.',data)
    # to transform \s. into .
    data=re.sub(r'\s\.','.',data)
    data=re.sub(r'[\"\']','',data)
    #data=re.sub(r'\.\s','> <',data)
    data=data.lower()
    data=remove_nonEnglish(data)
    with open("data_setFinal.txt",'w',encoding='utf8') as file1:
        file1.write(data)
        
    return data

data=preprocess()
```

```python
#Bigram model

from nltk.util import ngrams
from nltk import bigrams, trigrams
from collections import Counter, defaultdict, OrderedDict
bi_model = defaultdict(lambda: defaultdict(lambda: 0))
#data="water pollution and land pollution are two types of pollution. water pollution is contamination of water bodies. water pollution is dangerous to aquatic animals. Also water pollution is dangerous to human beings. "
def bigram_model():
    for sentence in data.split('.'):
        for w1, w2 in bigrams(sentence.split(), pad_right=True, pad_left=True):
            bi_model[w1][w2] += 1
     
  
    for w1 in bi_model: 
        total_count = float(sum(bi_model[w1].values()))
        for w2 in bi_model[w1]:
            bi_model[w1][w2] /= total_count
        bi_model[w1] =OrderedDict(sorted(bi_model[w1].items(), key=lambda x: x[1],reverse=True))

        
def getBigram(word):
    if(word not in bi_model.keys()):
        return None
    return list((dict(bi_model[word])).keys())[0]


bigram_model()
```

```python
#Trigram Model

tri_model = defaultdict(lambda: defaultdict(lambda: 0))
def trigram_model(): 
    for sentence in data.split('.'):
        for w1, w2, w3 in trigrams(sentence.split(), pad_right=True, pad_left=True):
            tri_model[(w1, w2)][w3] += 1
     
    for w1_w2 in tri_model:
        total_count = float(sum(tri_model[w1_w2].values()))
        for w3 in tri_model[w1_w2]:
            tri_model[w1_w2][w3] /= total_count
        tri_model[w1_w2] =OrderedDict(sorted(tri_model[w1_w2].items(), key=lambda x: x[1],reverse=True))

        
def getTrigram(w1,w2):
    if((w1,w2) not in tri_model.keys()):
        return None
    return list((dict(tri_model[w1,w2])).keys())[0]


trigram_model()
```

```python
#Quadgram Model

quad_model = defaultdict(lambda: defaultdict(lambda: 0))
def quadgram_model():
    for sentence in data.split('.'):
        for w1, w2, w3, w4 in list(ngrams(sentence.lower().split(),4)):
            quad_model[(w1, w2,w3)][w4] += 1
     
    for w1_w2_w3 in quad_model:
        total_count = float(sum(quad_model[w1_w2_w3].values()))
        for w4 in quad_model[w1_w2_w3]:
            quad_model[w1_w2_w3][w4] /= total_count
        quad_model[w1_w2_w3] =OrderedDict(sorted(quad_model[w1_w2_w3].items(), key=lambda x: x[1],reverse=True))


def getQuadgram(w1,w2,w3):
    if((w1,w2,w3) not in quad_model.keys()):
        return None
    return list((dict(quad_model[w1,w2,w3])).keys())[0] 


quadgram_model()
```
